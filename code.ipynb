Apro il file

```python
import csv

punti = []

with open('Edificio1.csv') as csvfile:
    reader = csv.reader(csvfile)
    next(reader)
    for row in reader:
        punti.append([elem for elem in row])
```

Creo vettori per ogni coordinata di ogni punto

```python
n = len(punti)
xp = []
for i in range(n):
    xp.append(float(punti[i][0]))

yp = []
for i in range(n):
    yp.append(float(punti[i][1]))

zp = []
for i in range(n):
    zp.append(float(punti[i][2]))
```

Costruisco il grafo con le distanze fattibili

```python
import math

def connessi(i, j):
    dx = abs(xp[i] - xp[j])
    dy = abs(yp[i] - yp[j])
    dz = abs(zp[i] - zp[j])
    dist = math.sqrt(dx**2 + dy**2 + dz**2) # SFRUTTIAMO PITAGORA PER CALCOLARE DISTANZA
    if dist <= 4:
        return True
    if dist <= 11 and sum([dx <= 0.5, dy <= 0.5, dz <= 0.5]) >= 1: #verifica se almeno una delle tre condizioni Ã¨ vera
        return True
    return False


N = {i: [] for i in range(len(punti))}
for i in range(n):
    for j in range(n):
         if i != j and connessi(i, j):
            N[i].append(j)
print(N)

    
```
